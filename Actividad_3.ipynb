```python
class Numero:
    def __init__(self,numero):
        self.normal = numero
        self.romano = ""
        self.convertir_a_romano()
        
    def convertir_a_romano(self):
        entero = self.normal
        numeros = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        numerales = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I']
        i = 0
        numeral = ""
        while entero > 0:
            for _ in range(entero // numeros[i]):
                numeral += numerales[i]
                entero -= numeros[i]
            i += 1
        self.romano = numeral
        
    def convertir_a_numero(self):
        romano = self.romano
        romanos = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 100}
        entero = 0
        for i in range(len(romano)):
            if i > 0 and romanos[romano[i]] > romanos[romano[i - 1]]:
                entero += romanos[romano[i]] - 2 * romanos[romano[i - 1]]
            else:
                entero += romanos[romano[i]]
            
        self.normal = entero
     
      
    def imprime(self):
        print(self.normal, self.romano)

        
    def suma_romano(self,romano):
        try:
            entero = self.normal
            self.romano = romano
            self.convertir_a_numero()
            self.normal = entero + self.normal
            self.convertir_a_romano()
        except:
            #esta captura de error se puede hacer tambien resta,multiplicacion,etc
            print("Ha fallado el número " + str(romano))
                    
    def is_romano(self,variable):
        try:
            romanos = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 100}
            for i in range(len(variable)):
                romanos[variable[i]]
        except:        
            return False
        else:
            return True
```
